```
(
df.groupby('param_marka-pojazdu')['price_value']
 .agg(np.mean)
 .sort_values(ascending=False)
 .head(50) #50 najdroższych
)  .plot(kind='bar', figsize=(15,5))
```

```
(
df.groupby('param_marka-pojazdu')['price_value']
 .agg(np.median)
 .sort_values(ascending=False)
 .head(50) #50 najdroższych
)  .plot(kind='bar', figsize=(15,5))
```

```
(
df.groupby('param_marka-pojazdu')['price_value']
 .agg([np.mean,np.median,np.size])
 .sort_values(by='mean', ascending=False)
 .head(50) #50 najdroższych
)  .plot(kind='bar', figsize=(20,5))
```

```
(
df.groupby('param_marka-pojazdu')['price_value']
 .agg([np.mean,np.median,np.size])
 .sort_values(by='size', ascending=False)
 .head(50) #50 najdroższych
)  .plot(kind='bar', figsize=(20,5), subplots=True)
```

```
def group_and_barplot(feat_groupby, feat_agg='price_value', agg_funcs=[np.mean,np.median,np.size], feat_sort ='mean', subplots=True):
  return (
  df
  .groupby(feat_groupby)[feat_agg]
  .agg(agg_funcs)
  .sort_values(by=feat_sort, ascending=False)
  .head(50) #50 najdroższych
  )  .plot(kind='bar', figsize=(20,5), subplots=subplots)
```

```
group_and_barplot('param_marka-pojazdu');
```

```
group_and_barplot('param_kraj-pochodzenia');
```

```
group_and_barplot('param_kolor', feat_sort='mean');
```
